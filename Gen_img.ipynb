```python
import cv2
import os

# Initialize video capture objects for each video file
videos = [
"D:\cpf\Videos\Basil Minced Beef Rice\RTE_Saraburi_ch4_20241011112856_20241011123152.mp4"
]
name = "Basil Minced Beef Rice"
caps = [cv2.VideoCapture(video) for video in videos]

# Create the output directory if it doesn't exist
output_dir = r"D:\cpf\code_for_gen_img\Img4TrainTiny"
os.makedirs(output_dir, exist_ok=True)

img_counter = 0

# Process each video
for i, cap in enumerate(caps):
    while True:
        ret, frame = cap.read()

        # If the end of the video is reached, break the loop
        if not ret:
            break

        # Show the current frame of the video
        cv2.imshow(f'Video {i+1}', frame)

        # Check for key press
        k = cv2.waitKey(1) & 0xFF

        # Save frames when 's' is pressed
        if k == ord('s'):
            img_name = os.path.join(output_dir, f"{name}_video{i+1}_frame_{img_counter}.png")
            cv2.imwrite(img_name, frame)
            print(f"Saved: {img_name}")
            img_counter += 1

        # Exit the loop if 'q' is pressed
        if k == ord('q'):
            break

    cap.release()

# Close all OpenCV windows
cv2.destroyAllWindows()
```
